```python
start = '2009-05-04'
end = '2024-12-27'

nasdaq_data = fdr.DataReader('IXIC', start, end)
print(nasdaq_data.head(10))
```

```python
print(nasdaq_data.info())
```

```python
print(nasdaq_data.describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(nasdaq_data.index, nasdaq_data['Close'], label='NASDAQ Index', color='blue')
plt.title('NASDAQ Index from 2009-05-04 to 2024-12-26')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.legend()
plt.grid()
plt.show()
```
